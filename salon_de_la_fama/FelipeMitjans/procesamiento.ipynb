```python
import pandas as pd
```

```python
datos = pd.read_csv("astronauts.csv")
datos = datos.drop("Group", 1)
```

```python
datos.columns
```

```python
for lugar in datos["Birth Place"]:
    if lugar[-1].islower():
        print(lugar.split(","))
```

```python
countries = []
for lugar in datos["Birth Place"]:
    if lugar[-1].islower():
        countries.append(lugar.split(",")[-1].strip())
    else:
        countries.append("US")
countries2 = []
for element in countries:
    if countries.count(element) > 1:
        countries2.append(element)
    else:
        countries2.append("Other")
datos["Country"] = countries2
```

```python
military = datos["Military Branch"].notnull()
datos["Military"] = military
datos[datos["Military"] == False]
```

```python
datos[datos["Death Mission"].notnull()].head()
```

```python
datos = datos.sort_values(by=['Name'])
datos.head()
```

```python
orden = []
for status in datos["Status"]:
    if status == "Deceased":
        orden.append(1)
    elif status == "Retired":
        orden.append(2)
    elif status == "Management":
        orden.append(3)
    elif status == "Active":
        orden.append(4)
    else:
        print("Error")
datos["orden"] = orden
```

```python
datos = datos.sort_values(by=["orden"])
datos = datos.drop("orden", 1)
```

```python
datos
```

```python
missions = set()
counter = 0
for row in datos[datos["Missions"].notnull()]["Missions"]:
    lista = row.split(",")
    for mision in lista:
        if "Discovery" in mision.strip():
            missions.add("Discovery")
        elif "Columbia" in mision.strip():
            missions.add("Columbia")
        elif "Atlantis" in mision.strip():
            missions.add("Atlantis")
        elif "Endeavor" in mision.strip():
            missions.add("Endeavor")
        elif "Soyuz" in mision.strip():
            missions.add("Soyuz")
        elif "Challenger" in mision.strip():
            missions.add("Challenger")
        elif "Skylab" in mision.strip():
            missions.add("Skylab")
        elif "Apollo" in mision.strip():
            missions.add("Apollo")
        elif "Gemini" in mision.strip():
            missions.add("Gemini")
        elif "Mercury" in mision.strip():
            missions.add("Mercury")
        else:
            missions.add(mision.strip())
print(missions)
```

```python
datos.to_csv("astronautas2.csv")
```
